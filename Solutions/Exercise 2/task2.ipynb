```python
def cor(X, Y):
    assert (len(X) == len(Y))

    meanX, meanY = sum(X) / len(X), sum(Y) / len(Y)

    X, Y = [x - meanX for x in X], [y - meanY for y in Y]
        
    num = sum(x * y for x, y in zip(X, Y))
    denom = (sum(x * x for x in X) * sum(y * y for y in Y)) ** .5
        
    return num / denom


def is_close(v1, v2): return abs(v1 - v2) < 1e-5


assert cor([1, 2, 3, 4, 5], [2, 4, 6, 8, 10]) == 1
assert cor([1, 2, 3, 4, 5], [10, 8, 6, 4, 2]) == -1
assert is_close(cor([1, 2, 3, 4, 5], [4, 1, 3, 5, 2]), 0)
assert cor([1, 2, 3, 4, 5, 6], [2, 3, 4, 5, 6, 7]) == 1
assert cor([10, 20, 30, 40, 50], [10, 19, 29, 35, 50]) > .9
```
