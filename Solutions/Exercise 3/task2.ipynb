```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.manifold import MDS
```

```python
data = pd.read_csv('data/eurodist.csv')

data
```

```python
mds = MDS(random_state=1, dissimilarity='precomputed')

coord = mds.fit_transform(data.iloc[:, 1:])

coord
```

```python
plt.figure(figsize=(12, 9))
plt.title('PCoA Result')

for (x, y), city in zip(coord, data.columns[1:]):
    plt.text(x, y, city, fontsize=9)

plt.scatter(coord[:, 0], coord[:, 1])
plt.show()
```
