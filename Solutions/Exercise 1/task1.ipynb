```python
with open(file) as f:
    txt = f.read()
    print(txt.count('Mathematics'))
```

```python
with open(file) as f:
    txt = f.read().lower()
    split = re.split('\W+', txt)

    count = {word: 0 for word in split}
    for word in split:
        count[word] += 1

    print(*(x[0] for x in sorted(count.items(), key=lambda x: x[1])[:-10:-1]))
```
