```python
# for the editor not to go crazy we simply declare them here
y_test, y_prob1, y_prob2 = [None] * 3
%store -r y_test
%store -r y_prob1
%store -r y_prob2
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
```

```python
pr1, rc1, _ = precision_recall_curve(y_test, y_prob1, pos_label='yes')
pr2, rc2, _ = precision_recall_curve(y_test, y_prob2, pos_label='yes')

plt.figure(figsize=(8, 6))

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')

plt.grid(True)

plt.plot(rc1, pr1, label=f'Classifier 1')
plt.plot(rc2, pr2, label=f'Classifier 2')

plt.legend()

plt.show()
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef

y_pred1 = np.where(y_prob1 >= 0.5, 'yes', 'no')
y_pred2 = np.where(y_prob2 >= 0.5, 'yes', 'no')

accuracy1 = accuracy_score(y_test, y_pred1)
accuracy2 = accuracy_score(y_test, y_pred2)
f11 = f1_score(y_test, y_pred1, pos_label='yes')
f12 = f1_score(y_test, y_pred2, pos_label='yes')
mcc1 = matthews_corrcoef(y_test, y_pred1)
mcc2 = matthews_corrcoef(y_test, y_pred2)

accuracy1, accuracy2, f11, f12, mcc1, mcc2, sum(y_test == 'yes')/len(y_test)
```
